```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from pylab import rcParams
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
#Data
run1_rate = [455,486,407,373,float('NaN')]
run2_rate = [360,283,305,314,264]

run1_pulse_messure = [60,70,67.5,65,66]
run1_pulse_watch = [72,76,82,72,74]
run2_pulse_messure = [72,67,62,65,61,64]
run2_pulse_watch = [64,64,72,70,65,64]

run1_pressure = [0.45,0.39,0.45,0.38,0.40]
run2_pressure = [0.41,0.44,0.43,0.44,0.43,0.39]

GPS=[4,0,0,0,0,2,3,2,3,6,2,5,9,6,6,8,9,2,0,0,8,6,5,0,5,2,0,0,0,0,2,3,3,4,5,5,6,6,6]
```

```python
#Latency
fig = plt.figure()
plt.plot(run1_rate, '-r', label="Run 1")
plt.plot(run2_rate, '-g', label="Run 2")
plt.xlabel('Measure', fontsize=18)
plt.ylabel('Time (in s)', fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=17)
rcParams['figure.figsize'] = 12, 5
plt.legend(loc='upper right',fontsize=18)
fig.savefig('latency_diff.jpg')
```

```python
#Heartbeat first run
fig = plt.figure()
plt.plot(run1_pulse_messure, '-r', label="Walker")
plt.plot(run1_pulse_watch, '-g', label="IWatch")
plt.xlabel('Sample', fontsize=18)
plt.ylabel('Pulse', fontsize=18)
rcParams['figure.figsize'] = 12, 5
plt.legend(loc='upper right',fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=17)
fig.savefig('pulse_run1_diff.jpg')
```

```python
#Heartbeat second run
fig = plt.figure()
plt.plot(run2_pulse_messure, '-r', label="Walker")
plt.plot(run2_pulse_watch, '-g', label="IWatch")
plt.xlabel('Sample', fontsize=18)
plt.ylabel('Pulse', fontsize=18)
rcParams['figure.figsize'] = 12, 5
plt.legend(loc='upper right',fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=17)
fig.savefig('pulse_run2_diff.jpg')
```

```python
#Pressure
fig = plt.figure()
plt.plot(run1_pressure, '-r', label='Run 1')
plt.plot(run2_pressure, '-g', label='Run 2')
plt.xlabel('Sample', fontsize=18)
plt.ylabel('Pulse', fontsize=18)
rcParams['figure.figsize'] = 12, 5
plt.legend(loc='upper right',fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=20)

fig.savefig('pressure_diff.jpg')
```
